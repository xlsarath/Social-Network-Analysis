## 7 The dispersion between the two nodes with highest PageRank

```python
PAGE_RANK =nx.pagerank(G)
```

```python
def top_five(ranks):
    """gets the top-five blogs by rank"""
    top = list(reversed(sorted((rank, node)
                               for node, rank in ranks.items())))[:5]
    return [node for rank, node in top]
```

```python
def answer_seven():
    """Top 5 nodes by page rank"""
    return top_five(PAGE_RANK)
```

```python
print(answer_seven())
```

## 9.Find the communities using the Girvan Newman algorithm. Print the number of communities

```python
from networkx.algorithms import community # community lib isn't part of nx so importing it separately
communities_generator = community.girvan_newman(G)
```

### to print list of communities using girvan_newman

```python
tuple(sorted(c) for c in next(communities_generator))
```

## 10 .(cont'd from 9) Print the top-5 nodes for each community you found in step 9.

```python
import itertools
k = 5
comp = community.girvan_newman(G)
for communities in itertools.islice(comp, k):
    print(tuple(sorted(c) for c in communities)) 
```
